```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Vertices
p1 = np.array([0,0])
p2 = np.array([2,0])
p3 = np.array([1,2])

vertices = [p1, p2, p3]
puntos = [p1, p2, p3]

# Punto de inicio
x = random.uniform(0, 2)
if x<=1:
  y = random.uniform(0, 2*x)
  inicio = np.array([x,y])
else:
  y = random.uniform(0, -2*x+4)
  inicio = np.array([x,y]) 
puntos.append(inicio)

#Defino dominio e imagen de los puntos
dom = [point[0] for point in puntos]
im = [point[1] for point in puntos]
# Grafico
plt.scatter(dom, im)
plt.grid()
plt.show()
```

```python
modulo1_inicio = np.linalg.norm(p1-inicio)
modulo2_inicio = np.linalg.norm(p2-inicio)
modulo3_inicio = np.linalg.norm(p3-inicio)

# Elijo el minimo de los modulos para realizar el primer punto del algoritmo
modulos = [modulo1_inicio, modulo2_inicio, modulo3_inicio]
if min(modulos)== modulo1_inicio:
  p = (p1+inicio)/2
  puntos.append(p)
elif min(modulos)== modulo2_inicio:
  p = (p2+inicio)/2
  puntos.append(p)
else:
  p = (p3+inicio)/2
  puntos.append(p)

#Defino dominio e imagen de los puntos
dom = [point[0] for point in puntos]
im = [point[1] for point in puntos]
# Grafico
plt.scatter(dom, im)
plt.grid()
plt.show()
```

```python
# Loopeo diezmil veces para hallar el patron
r=0
while r < 10000:
  vertice = random.choice(vertices)
  p = (vertice + p)/2
  puntos.append(p)
  # Incremento r
  r+=1

#Defino dominio e imagen de los puntos
dom = [point[0] for point in puntos]
im = [point[1] for point in puntos]
# Grafico Final
plt.figure(figsize=[10,10])
plt.scatter(dom, im)
plt.grid()
plt.show()
```
