### Importo y defino la funcion para graficar y extraer las columnas como listas

```python
import numpy as np
import matplotlib.pyplot as plt

def graf(data, titulo):
  datos = np.loadtxt(data)

  t = []
  vol = []

  for i in range(0,len(datos)):
    t.append(datos[i][0])
    vol.append(datos[i][1])

  # Datos
  plt.figure(figsize=[30,10])
  plt.scatter(t, vol, c='black')

  plt.title(titulo, fontsize=25)
  plt.xlabel('Tiempo [s]', fontsize=15)
  plt.ylabel('Voltaje [V]', fontsize=15)
  plt.grid()
  plt.show()

  return t, vol
```

### Grafico los canales

```python
t1 , vol1 = graf("CH1god.txt", "Canal 1")
```

```python
t2 , vol2 = graf("CH2god.txt", "Canal 2")
```

### Grafico los batidos

```python
# Esto se puede hacer porque la longitud de vol1 (len(vol1)) y vol2 son iguales
vol1 = np.array(vol1)
vol2 = np.array(vol2)

vol_onda = vol1 * vol2

# Defino un 'dominio' para graficar vol_onda entre el valor minimo de t1 y el valor max de t1, con la longitud de t1
X = np.linspace(t1[0],t1[-1], len(t1))

# Grafico todo junto
plt.figure(figsize=[30,10])

plt.scatter(X, vol_onda, c='red',label='Batido loco')
# Multiplico por 100 los vol para expandir el eje 'y' para las ondas, porq si no no se nota nada
plt.scatter(t1, 100*vol1, c='blue',label='Datos del canal 1 x100')
plt.scatter(t2, 100*vol2, c='yellow',label='Datos del canal 2 x100')


plt.xlabel('Tiempo [s]', fontsize=15)
plt.ylabel('Voltaje [V]', fontsize=15)
plt.legend(loc=2, fontsize=15)
plt.grid()
plt.show()
```

```python
# Grafico todo sin escalar para que vean como se ve
plt.figure(figsize=[30,10])
plt.scatter(X, vol_onda, c='red',label='Batido loco')
plt.scatter(t1,vol1, c='blue',label='Datos del canal 1')
plt.scatter(t2,vol2, c='yellow',label='Datos del canal 2')
plt.xlabel('Tiempo [s]', fontsize=15)
plt.ylabel('Voltaje [V]', fontsize=15)
plt.legend(loc=2, fontsize=15)
plt.grid()
plt.show()
```
